# Recipes

```python
# Menampilkan total baris
total_rows = df_recipes.shape[0]
print(f"Total baris dalam dataset: {total_rows}")
```

```python
df_recipes.info()
```

```python
# Tampilkan baris di mana 'RecipeIngredientQuantities' mengandung huruf NA
df_with_na = df_recipes[df_recipes['RecipeIngredientQuantities'].apply(lambda x: pd.isna(x) or 'NA' in str(x))]

print(df_with_na.shape[0])
```

```python
# Hapus baris di mana 'RecipeIngredientQuantities' mengandung huruf NA
df_recipes = df_recipes.drop(df_with_na.index)

df_recipes.head(5)
```

```python
# Menghapus baris yang tidak memiliki link gambar
df_recipes = df_recipes[df_recipes['Images'].notna()]

# Menghapus baris dengan nilai "(0)" atau karakter kosong pada kolom "Images"
df_recipes = df_recipes[df_recipes['Images'] != 'character(0)']

# Menghapus baris yang memiliki ReviewCount di bawah 10
df_recipes = df_recipes[df_recipes['ReviewCount'] >= 2]

# Menghapus kolom-kolom yang diminta
columns_to_drop = ['AuthorName', 'CookTime', 'PrepTime', 'TotalTime', 'DatePublished', 'RecipeYield', 'Keywords', 'RecipeServings']
df_recipes = df_recipes.drop(columns=columns_to_drop, errors='ignore')
```

```python
df_recipes = df_recipes.rename(columns={
    'FatContent': 'Fat',
    'SaturatedFatContent': 'SaturatedFat',
    'CholesterolContent': 'Cholesterol',
    'SodiumContent': 'Sodium',
    'CarbohydrateContent': 'Carbohydrate',
    'FiberContent': 'Fiber',
    'SugarContent': 'Sugar',
    'ProteinContent': 'Protein'
})
```

```python
df_recipes['Description'] = df_recipes['Description'].replace('from Food.com', '', regex=True)
```

```python
# Menampilkan nama yang duplikat
duplicate_names = df_recipes[df_recipes.duplicated(subset='Name', keep=False)]['Name'].unique()

# Menampilkan hasil
print("Nama yang duplikat:")
print(duplicate_names)
```

```python
# Menghapus baris dengan nama yang duplikat
df_recipes = df_recipes.drop_duplicates(subset='Name', keep='first')
```

```python
#check
check_rows = df_recipes.loc[df_recipes['Name'] == "Southern Buttermilk Biscuits"]
check_rows.head()
```

```python
# Menghapus karakter backslash (\) dan kutip ganda (") setelah backslash
df_recipes['RecipeInstructions'] = df_recipes['RecipeInstructions'].str.replace(r'\\(")', r'')
```

```python
def map_str_to_list(value):
    if isinstance(value, str):
        pattern = re.compile(r'\"(.*?)\"')
        return pattern.findall(value)
    else:
        return value

def map_for_series(series):
    return series.apply(map_str_to_list)
```

```python
modify_col_list = ['RecipeIngredientQuantities', 'RecipeIngredientParts', 'RecipeInstructions']
df_recipes = df_recipes.apply(lambda x: map_for_series(x) if x.name in modify_col_list else x)
```

```python
df_recipes['Ingredients'] = df_recipes.apply(
    lambda row: [f"{row['RecipeId']} \"{ingredient}\"" for ingredient in row['RecipeIngredientParts']],
    axis=1
)
```

```python
#Menggabungkan
df_recipes['Ingredients'] = df_recipes.apply(lambda row: [f'{quantity} "{ingredient}"' for quantity, ingredient in zip(row['RecipeIngredientQuantities'], row['RecipeIngredientParts'])], axis=1)
```

```python
df_recipes['Instructions'] = df_recipes['RecipeInstructions'].apply(lambda x: [f'"{step}"' for step in x])
```

```python
# Menghapus kolom-kolom yang diminta
columns_drop = ['RecipeInstructions', 'RecipeIngredientQuantities', 'RecipeIngredientParts']
df_recipes = df_recipes.drop(columns=columns_drop, errors='ignore')
```

```python
df_recipes = df_recipes[~df_recipes['Name'].str.contains('pork', case=False)]
```
